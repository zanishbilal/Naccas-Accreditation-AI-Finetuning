```python
# âœ… Imports
import os
import json
import torch, gc
import pdfplumber
# import shutil
from datasets import Dataset
from transformers import (
    AutoModelForCausalLM,
    AutoTokenizer,
    TrainingArguments,
    BitsAndBytesConfig,
    DataCollatorForLanguageModeling,
    Trainer,
    get_scheduler

)
from peft import prepare_model_for_kbit_training, get_peft_model, LoraConfig, TaskType
```

```python

# âœ… PDF Preprocessing
def preprocess_pdf(pdf_path):
    text = ""
    with pdfplumber.open(pdf_path) as pdf:
        for page in pdf.pages:
            page_text = page.extract_text()
            if page_text:
                text += page_text + "\n"
    return text.strip()

# âœ… Chunking
def chunk_text(text, tokenizer, max_tokens=512, stride=64):
    input_ids = tokenizer(text, return_tensors='pt')['input_ids'][0]
    chunks = []
    for i in range(0, len(input_ids), max_tokens - stride):
        chunk_ids = input_ids[i:i + max_tokens]
        chunk = tokenizer.decode(chunk_ids, skip_special_tokens=True)
        chunks.append(chunk.strip())
    return chunks

# âœ… File paths
def get_file_paths(folder_path):
    file_paths = []
    for root, _, files in os.walk(folder_path):
        for file in files:
            file_paths.append(os.path.join(root, file))
    return file_paths

```

```python
folder_path = "/content/drive/MyDrive/Clara Project/Clara docs"
file_paths = get_file_paths(folder_path)
model_name = "meta-llama/Llama-3.2-3B"
output_dir = "/content/drive/MyDrive/llama3_policy_finetune_v2.5"
```

```python
# âœ… Clean VRAM
gc.collect()
torch.cuda.empty_cache()

# âœ… Tokenizer
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

# âœ… Preprocess PDFs and chunk
all_chunks = []
for path in file_paths:
    text = preprocess_pdf(path)
    chunks = chunk_text(text, tokenizer)
    all_chunks.extend([{"text": chunk} for chunk in chunks])

dataset = Dataset.from_list(all_chunks)

# âœ… Tokenization
def tokenize_function(example):
    return tokenizer(example["text"], padding="max_length", truncation=True, max_length=512)

tokenized_dataset = dataset.map(tokenize_function, batched=True)

# âœ… Quantization config
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4"
)

# âœ… Data collator
data_collator = DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False)

# âœ… LoRA config
lora_config = LoraConfig(
    r=32,
    lora_alpha=16,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],
    lora_dropout=0.05,
    bias="none",
    task_type=TaskType.CAUSAL_LM
)
```

```python

# âœ… Load Model
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=bnb_config,
    device_map="auto",
    trust_remote_code=True
)
model = prepare_model_for_kbit_training(model)
model = get_peft_model(model, lora_config)

# âœ… Training args
training_args = TrainingArguments(
    output_dir=output_dir,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=4,
    gradient_checkpointing=True,
    learning_rate=2e-4,
    weight_decay=0.01,
    num_train_epochs=50,
    logging_steps=10,
    save_strategy="no",
    save_total_limit=1,
    fp16=True,
    report_to="wandb",
    max_grad_norm=0.5,
    warmup_steps=100
)

# âœ… Optimizer and LR scheduler
optimizer = torch.optim.AdamW(model.parameters(), lr=2e-4, weight_decay=0.01)
lr_scheduler = get_scheduler(
    name="linear",
    optimizer=optimizer,
    num_warmup_steps=100,
    num_training_steps=(len(tokenized_dataset) // training_args.per_device_train_batch_size * training_args.num_train_epochs)
)

# âœ… Trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    data_collator=data_collator,
    optimizers=(optimizer, lr_scheduler)
)

# âœ… Train
trainer.train()

# âœ… Save final model to Drive
model.save_pretrained(output_dir)
tokenizer.save_pretrained(output_dir)

print(f"âœ… Final model saved to Drive: {output_dir}")
```

```python
import os, json
from unsloth import FastLanguageModel
from transformers import TrainingArguments, DataCollatorForLanguageModeling, AutoTokenizer
from trl import SFTTrainer
from datasets import Dataset
import torch


# Load the model
model, tokenizer = FastLanguageModel.from_pretrained(
    model_name="unsloth/Llama-3.2-3B-Instruct-bnb-4bit",  # Pre-quantized 4-bit model
    load_in_4bit=True,  # Enable 4-bit quantization
    max_seq_length=2048,
    device_map="auto",  # Automatically map to available devices
)

# Add dropout (optional, uncomment if needed)
# model.config.hidden_dropout_prob = 0.1  # Optional extra dropout
# model.config.attention_dropout = 0.1    # Add dropout in attention if desired

# Prepare model for QLoRA
model = FastLanguageModel.get_peft_model(
    model,
    r=8,
    lora_alpha=16,
    lora_dropout=0.0,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj", "gate_proj", "up_proj", "down_proj"],  # Apply to all modules
    use_gradient_checkpointing='unsloth',
    random_state=42,
    use_rslora=False,
)
```

```python
# Path to the directory containing your 17 JSON files
json_dir = "/content/drive/MyDrive/Carla instruction pairs json "

data = []
for filename in os.listdir(json_dir):
    if filename.endswith(".json"):
        with open(os.path.join(json_dir, filename), "r", encoding="utf-8") as f:
            data.extend(json.load(f))  # assume each JSON file is a list of dicts

# Optionally check one example
print(data[0])

# Convert to Hugging Face dataset
dataset = Dataset.from_list(data)

split_dataset = dataset.train_test_split(test_size=0.2, seed=42)

train_dataset = split_dataset["train"]
val_dataset = split_dataset["test"]

def format_example(example):
    if "instruction" not in example or "response" not in example:
        raise ValueError(f"Invalid example: {example}")
    return {
        "text": f"<s>[INST] {example['instruction']} [/INST] {example['response']}</s>"
    }

train_dataset = train_dataset.map(format_example)
val_dataset = val_dataset.map(format_example)

tokenizer = AutoTokenizer.from_pretrained("unsloth/Llama-3.2-3B-Instruct-bnb-4bit")

def tokenize(example):
    return tokenizer(example["text"], truncation=True, max_length=2048)

train_dataset = train_dataset.map(tokenize)
val_dataset = val_dataset.map(tokenize)
```

```python

# Training Args
args = TrainingArguments(
    output_dir = "/content/v3_llama_3.2_instruct",     # ğŸ‘ˆ Clean folder name
    num_train_epochs = 30,
    per_device_train_batch_size = 8,
    gradient_accumulation_steps = 8,
    save_strategy = "epoch",                           # Save at every epoch
    eval_strategy = "epoch",                     # Evaluate at every epoch
    save_total_limit = 1,                              # Only keep the best model
    load_best_model_at_end = True,                     # ğŸ‘ˆ Save best based on metric
    metric_for_best_model = "eval_loss",               # ğŸ‘ˆ Use lowest validation loss
    greater_is_better = False,                         # Lower eval_loss is better
    logging_steps = 10,
    learning_rate = 2e-5,
    lr_scheduler_type = "cosine",
    warmup_steps = 20,
    weight_decay = 0.01,
    fp16 = True,
    report_to = "wandb"
)



trainer = SFTTrainer(
    model=model,
    tokenizer=tokenizer,
    train_dataset=train_dataset,
    eval_dataset=val_dataset,
    args=args,
    max_seq_length=2048,
    data_collator=DataCollatorForLanguageModeling(tokenizer=tokenizer, mlm=False),
)
```

```python
# Start training
trainer.train()

# Save best model manually to /content
trainer.save_model("/content/llama3.2-instruct-best")
```
